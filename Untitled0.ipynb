```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

```

```python
# Load the dataset
df = pd.read_csv('co2_emissions.csv', header=[0, 1])

# Display the first few rows to inspect its structure
df.head()
```

```python
# Rename columns
df.columns = [''.join(col).strip() for col in df.columns.values]

# Display the first few rows after renaming
df.head()
```

```python
# Rename columns to appropriate lables
corrected_columns = {
    'CountryWorld' : 'Country',
    '1990 22674.120':'1990',
    '2005 30049.810' : '2005',
    '2017 37077.400' : '2017',
    '2021 37857.580': '2021'
    }
df.rename(columns=corrected_columns, inplace=True)

# Remove non-country rows
# We can identify non-country rows by checking if the 'Country' name contains "World"
df_countries = df[~df['Country'].str.contains("World")]

# Display the cleaded dataframe
df_countries.head()
```
